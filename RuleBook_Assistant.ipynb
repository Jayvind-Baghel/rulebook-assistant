```python
# --- Set File Path ---
pdf_path = "/content/drive/MyDrive/Rulebook.pdf"  # Update path if needed

# --- Extract Text from PDF ---
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = "\n".join([page.get_text("text") for page in doc])
    return text.strip()

pdf_text = extract_text_from_pdf(pdf_path)
print(pdf_text[:1000])  # Preview first 1000 characters

# --- Clean Text ---
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

cleaned_text = clean_text(pdf_text)

# --- Chunk Cleaned Text ---
def chunk_text(text, chunk_size=500):
    splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=50)
    return splitter.split_text(text)

text_chunks = chunk_text(cleaned_text)

# --- Extract and Clean Tables from PDF ---
def clean_table_from_pdf(pdf_path):
    cleaned_rows = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            tables = page.extract_tables()
            for table in tables:
                current_row = []
                for row in table:
                    if not row or all(cell is None or cell.strip() == "" for cell in row):
                        continue
                    row = [cell.replace('\n', ' ').strip() if cell else "" for cell in row]

                    if len(current_row) > 0 and len(current_row) + len(row) <= 11:
                        current_row = [f"{a} {b}".strip() for a, b in zip(current_row + [""]*(11 - len(current_row)), row)]
                    elif len(row) == 11 and re.match(r"^\d+$", row[0]):
                        if current_row:
                            cleaned_rows.append(current_row)
                        current_row = row
                    else:
                        current_row = [f"{a} {b}".strip() for a, b in zip(current_row + [""]*(11 - len(current_row)), row)]

                if current_row and len(current_row) == 11:
                    cleaned_rows.append(current_row)

    columns = ["S.No", "Event", "P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "Venue"]
    return pd.DataFrame(cleaned_rows, columns=columns)

# --- Convert Table Rows to Text Chunks ---
def table_to_text_chunks(df):
    chunks = []
    for _, row in df.iterrows():
        event_name = row["Event"]
        points = ", ".join([f"{pos}: {row[pos]}" for pos in ["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8"]])
        venue = row["Venue"]
        chunk = f"Event: {event_name}\nPoints Distribution → {points}\nVenue: {venue}"
        chunks.append(chunk)
    return chunks

# --- Table Processing ---
table_df = clean_table_from_pdf(pdf_path)
table_chunks = table_to_text_chunks(table_df)
text_chunks.extend(table_chunks)

# --- Load Sentence Embedding Model ---
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# --- Initialize ChromaDB ---
chroma_client = chromadb.PersistentClient(path="/content/chromadb_store")
collection = chroma_client.get_or_create_collection(name="rulebook_data")

# --- Store Embeddings ---
for i, chunk in enumerate(text_chunks):
    embedding = embedding_model.encode(chunk)
    collection.add(ids=[str(i)], embeddings=[embedding], metadatas=[{"text": chunk}])

print(f"Stored {len(text_chunks)} text chunks in ChromaDB.")

# --- HuggingFace Login ---
login(token="Your_Hugging_Face_token")  # 🔐 Keep token secret if sharing code

# --- Load LLaMA-2 Model ---
model_name = "meta-llama/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    offload_folder="offload_weights"
)
print("Model loaded successfully!")

# --- Semantic Search Function ---
def search_relevant_text(query, k=3):
    query_embedding = embedding_model.encode(query)
    results = collection.query(query_embeddings=[query_embedding], n_results=k)
    return [item["text"] for item in results["metadatas"][0]]

# --- Generate Answer + Follow-ups ---
def generate_answer_with_followups(query):
    context = " ".join(search_relevant_text(query))
    base_prompt = f"""
You are a helpful assistant for answering questions about the VARCHASVA 25 Rulebook.

Use the context below to answer the question clearly and factually. Then, generate exactly three follow-up questions that would naturally come next based on the answer.

Separate the answer and the follow-up section with the heading:
"Follow-up Questions:"

Context:
{context}

Question: {query}

Answer:
"""
    inputs = tokenizer(base_prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=300)
    full_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return full_output.strip()
```

```python
# --- Start Interactive Q&A Loop ---
print("\n🔍 Ask anything about the VARCHASVA 25 Rulebook!")
print("Type 'q', 'quit', 'exit', or 'goodbye' to end.\n")

while True:
    user_query = input("❓ Your Question: ").strip().lower()

    if user_query in {"q", "quit", "exit", "goodbye"}:
        print("👋 Goodbye! See you next time.")
        break

    # Get original query for proper casing
    answer = generate_answer_with_followups(user_query)
    print("\n💬 Answer:\n")
    print(answer)
    print("\n" + "-"*80 + "\n")
```
